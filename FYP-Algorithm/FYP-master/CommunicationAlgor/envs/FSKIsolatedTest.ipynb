```python
SNR=30

plot=1

# Number of symbols
Ns = 256

# Intermediary frequency
Fsk = [2e3,3e3]

# Samples per second
Fs = 1024
```

### Data bitstream generation

```python
data = QAM.dataGenerator(int(Fs)) #  Total Bits count
if plot == 1:
    plt.figure(0)
    plt.stem(data)
    plt.title('Data')
    plt.grid()
```

#### Time sequence

```python
t = np.arange(start=0, stop=1, step=1/Fs)

```

```python

DataFSK=[]

for i in range(len(data)):
    #f.append(Fsk[int(i)])
    DataFSK.append(np.sin(pi*Fsk[int(data[i])]*t[i]/180.))
```

```python
figure, axis = plt.subplots(2)
axis[0].plot(t, data)
axis[0].set_title("Binary digital data")
axis[1].plot(t, DataFSK, 'r')
axis[1].set_title("FSK modulated signal")
plt.tight_layout()
plt.show()
```
