### Import needed Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
import math,cmath
import scipy.fftpack as sf
import statistics
```

### Import QAM functions from QAM

```python
import import_ipynb
import QAM
```

### Parameters for the Model

```python
SNR=30
plot=1

# Upsampler Factor
K = 10

# Number of symbols
Ns = 256

# Roll-off factor
alpha = 0.3

# Bits per symbol
Bs = 4

# 16QAM Constellation vector
QAM16 = [-1, -0.333, 0.333, 1]

# Intermediary frequency
Fif = 2e6

# Sampling Rate
Fs = Fif*K/2
```

### Data bitstream generation

```python
data = QAM.dataGenerator(Ns*Bs) #  Total Bits count
if plot == 1:
    plt.figure(0)
    plt.stem(data)
    plt.title('Data')
    plt.grid()
```

### Bit Slicing for Odd and Even

```python

(dataIn, dataQu) = QAM.Slice(data)
```

### Modulation

```python
mapI = QAM.Modulator16QAM(QAM16, dataIn)
mapQ = QAM.Modulator16QAM(QAM16, dataQu)


if plot == 1:
    plt.figure(1)
    plt.subplot(2, 1, 1)
    plt.stem(mapI)
    plt.title('Mapper I')
    plt.grid()
    plt.subplot(2, 1, 2)
    plt.stem(mapQ)
    plt.title('Mapper Q')
    plt.tight_layout()
    plt.grid()

    plt.figure(2)
    plt.scatter(mapI, mapQ)
    plt.title('Constelation IQ out mapper')
    plt.xlabel('In-Phase')
    plt.ylabel('Quadrature')
    plt.tight_layout()
    plt.grid()
```

### Upsammpling

```python
upI = QAM.Upsampler(Ns, K, mapI)
upQ = QAM.Upsampler(Ns, K, mapQ)


if plot == 1:
    plt.figure(3)
    plt.subplot(2, 1, 1)
    plt.stem(upI)
    plt.grid()
    plt.title('Up-Sampler I')
    plt.subplot(2, 1, 2)
    plt.stem(upQ)
    plt.title('Up-Sampler Q')
    plt.tight_layout()
    plt.grid()
```

### Pulse shapping

```python
[shape_I, a, b] = QAM.PulseShapping(upI, Ns, alpha, Fif, Fs)
[shape_Q, a, b] = QAM.PulseShapping(upQ, Ns, alpha, Fif, Fs)

template = b

if plot == 1:
    plt.figure(4)
    plt.plot(a, b)
    plt.title('SRRC Filter Impulse Response')
    plt.grid()
    plt.figure(5)
    plt.subplot(3, 1, 1)
    plt.plot(shape_I)
    plt.title('Raised Cosine Filter Convolution I')
    plt.grid()
    plt.subplot(3, 1, 2)
    plt.plot(shape_Q)
    plt.title('Raised Cosine Filter Convolution Q')
    plt.grid()
    plt.subplot(3, 1, 3)
    plt.plot(shape_I, shape_Q)
    plt.title('Constalation IQ filter output')
    plt.tight_layout()
    plt.grid()
```

```python

if plot == 1:
    X_f = abs(sf.fft(upI))
    l = np.size(upI)
    fr = (Fs/2)*np.linspace(0, 1, int(l/2))
    xl_m = (2/l)*abs(X_f[0:np.size(fr)])

    plt.figure(6)
    plt.subplot(2, 1, 1)
    plt.plot(fr/1e6, 20*np.log10(xl_m))
    plt.title('Upsampler Output Spectrum')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()
    plt.tight_layout()

    X_f2 = abs(sf.fft(shape_I))
    l2 = np.size(shape_I)
    fr2 = (Fs/2)*np.linspace(0, 1, int(l2/2))
    xl_m2 = (2/l2)*abs(X_f2[0:np.size(fr2)])

    plt.subplot(2, 1, 2)
    plt.plot(fr2/1e6, 20*np.log10(xl_m2))
    plt.title('Shaping Filter Output Spectrum')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()
    plt.tight_layout()
```

```python
# Oscillator

delta_phase = np.random.normal(0, pi/3, 1)
delta_freq = np.random.normal(0, 20, 1)

(loCos_TX, t) = QAM.CarrierGen(0, 4e-4, 2 /
                                (Fif*K),  Fif + delta_freq, delta_phase + pi/2)
(loSin_TX, t) = QAM.CarrierGen(0, 4e-4, 2 /
                                (Fif*K),  Fif + delta_freq, delta_phase)

if plot == 1:
    X_f_1 = abs(sf.fft(loCos_TX))
    l_1 = np.size(loCos_TX)
    fr_1 = (Fs/2)*np.linspace(0, 1, int(l_1/2))
    xl_m_1 = (2/l_1)*abs(X_f_1[0:np.size(fr_1)])

    plt.figure(7)
    plt.subplot(2, 1, 1)
    plt.plot(fr_1/1e6, 20*np.log10(xl_m_1))
    plt.title('Spectrum of local oscilator')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(t, loCos_TX)
    plt.title('Local Cos')
    plt.xlabel('t(s)')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()
```

### Mixer

```python
# Mixers

mixI = QAM.Mixer(shape_I, loCos_TX)
mixQ = QAM.Mixer(shape_Q,loSin_TX)

if plot == 1:
    plt.figure(8)
    plt.subplot(3, 1, 1)
    plt.plot(mixI)
    plt.title('Mix I')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.plot(mixQ)
    plt.title('Mix Q')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()
    
    seva = abs(sf.fft(mixI))
    nova = np.size(mixI)
    balti = (Fs/2)*np.linspace(0, 1, int(nova/2))
    druni = (2/nova)*abs(seva[0:np.size(balti)])
    
    plt.figure()
    plt.plot(balti/1e6, 20*np.log10(druni))
    plt.title('Spectrum of Carrier agregated')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()

```

### Combiner

```python
# Combiner

IF = QAM.combiner(mixI, mixQ)

# print(IF)

# with open("TransmitedData.txt", "w", newline="") as file:
#     for i in range(len(IF)):
#         file.write(str(IF[i]))


seva = abs(sf.fft(IF))
nova = np.size(IF)
balti = (Fs/2)*np.linspace(0, 1, int(nova/2))
druni = (2/nova)*abs(seva[0:np.size(balti)])

plt.figure()
plt.plot(balti/1e6, 20*np.log10(druni))
plt.title('Spectrum of Carrier agregated')
plt.xlabel('Frequency(MHz)')
plt.ylabel('Magnitute(dB)')
plt.grid()

np.save('OutputFiles/TransmittedData', IF)
```

```python
# Upsampler Factor
Ki = 10

# Number of symbols
Nsi = 256

# Roll-off factor
alphai = 0.3

# Bits per symbol
Bsi = 4

# 16QAM Constellation vector
QAM16 = [-1, -0.333, 0.333, 1]

# Intermediary frequency
Fifi = 2e6

Fci = 4e6

# Sampling Rate
Fsi = Fifi*Ki/2

Fcsi = Fci*Ki/2
```

```python
upIi = QAM.Upsampler(Nsi, Ki, mapIi)
upQi = QAM.Upsampler(Nsi, Ki, mapQi)


if plot == 1:
    plt.figure(3)
    plt.subplot(2, 1, 1)
    plt.stem(upIi)
    plt.grid()
    plt.title('Up-Sampler I')
    plt.subplot(2, 1, 2)
    plt.stem(upQi)
    plt.title('Up-Sampler Q')
    plt.tight_layout()
    plt.grid()
```

```python
[shape_Ii, ai, bi] = QAM.PulseShapping(upIi, Nsi, alpha, Fifi, Fsi)
[shape_Qi, ai, bi] = QAM.PulseShapping(upQi, Nsi, alpha, Fifi, Fsi)

templateI = bi

if plot == 1:
    plt.figure(4)
    plt.plot(ai, bi)
    plt.title('SRRC Filter Impulse Response')
    plt.grid()
    plt.figure(5)
    plt.subplot(3, 1, 1)
    plt.plot(shape_Ii)
    plt.title('Raised Cosine Filter Convolution I')
    plt.grid()
    plt.subplot(3, 1, 2)
    plt.plot(shape_Qi)
    plt.title('Raised Cosine Filter Convolution Q')
    plt.grid()
    plt.subplot(3, 1, 3)
    plt.plot(shape_Ii, shape_Qi)
    plt.title('Constalation IQ filter output')
    plt.tight_layout()
    plt.grid()
```

```python
if plot == 1:
    X_fi = abs(sf.fft(upIi))
    li = np.size(upIi)
    fri = (Fsi/2)*np.linspace(0, 1, int(li/2))
    xl_mi = (2/li)*abs(X_fi[0:np.size(fri)])

    plt.figure(6)
    plt.subplot(2, 1, 1)
    plt.plot(fri/1e6, 20*np.log10(xl_mi))
    plt.title('Upsampler Output Spectrum')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()
    plt.tight_layout()

    X_f2i = abs(sf.fft(shape_Ii))
    l2i = np.size(shape_Ii)
    fr2i = (Fsi/2)*np.linspace(0, 1, int(l2i/2))
    xl_m2i = (2/l2i)*abs(X_f2i[0:np.size(fr2i)])

    plt.subplot(2, 1, 2)
    plt.plot(fr2i/1e6, 20*np.log10(xl_m2i))
    plt.title('Shaping Filter Output Spectrum')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()
    plt.tight_layout()
```

```python
# Oscillator

delta_phasei = np.random.normal(0, pi/3, 1)
delta_freqi = np.random.normal(0, 20, 1)

(loCos_TXi, ti) = QAM.CarrierGen(0, 4e-4, 2 /
                                (Fci*Ki),  Fci + delta_freqi, delta_phasei + pi/2)
(loSin_TXi, ti) = QAM.CarrierGen(0, 4e-4, 2 /
                                (Fci*Ki),  Fci + delta_freqi, delta_phasei)

if plot == 1:
    X_f_1i = abs(sf.fft(loCos_TXi))
    l_1i = np.size(loCos_TXi)
    fr_1i = (Fcsi/2)*np.linspace(0, 1, int(l_1i/2))
    xl_m_1i = (2/l_1i)*abs(X_f_1i[0:np.size(fr_1i)])

    plt.figure(7)
    plt.subplot(2, 1, 1)
    plt.plot(fr_1i/1e6, 20*np.log10(xl_m_1i))
    plt.title('Spectrum of local oscilator')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(ti, loCos_TXi)
    plt.title('Local Cos')
    plt.xlabel('t(s)')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()
```

```python
# Mixers

mixIi = QAM.Mixer(shape_Ii, loCos_TXi)
mixQi = QAM.Mixer(shape_Qi,loSin_TXi)

if plot == 1:
    plt.figure(8)
    plt.subplot(3, 1, 1)
    plt.plot(mixIi)
    plt.title('Mix I')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.plot(mixQi)
    plt.title('Mix Q')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()
    
    sevai = abs(sf.fft(mixIi))
    novai = np.size(mixIi)
    baltii = (Fcsi/2)*np.linspace(0, 1, int(novai/2))
    drunii = (2/novai)*abs(sevai[0:np.size(baltii)])
    
    plt.figure()
    plt.plot(baltii/1e6, 20*np.log10(drunii))
    plt.title('Spectrum of Carrier agregated')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()

```

```python
# Combiner

IFi = QAM.combiner(mixIi, mixQi)

sevai = abs(sf.fft(IFi))
novai = np.size(IFi)
baltii = (Fcsi/2)*np.linspace(0, 1, int(novai/2))
drunii = (2/novai)*abs(sevai[0:np.size(baltii)])

plt.figure()
plt.plot(baltii/1e6, 20*np.log10(drunii))
plt.title('Spectrum of Carrier agregated')
plt.xlabel('Frequency(MHz)')
plt.ylabel('Magnitute(dB)')
plt.grid()

np.save('OutputFiles/TransmittedData', IFi)
```

### Mixer

```python
# Mixer

shape_I_demod = QAM.Mixer(IFg, loCos_TX)
shape_Q_demod = QAM.Mixer(IFg, loSin_TX)

if plot == 1:
    plt.figure(12)
    plt.subplot(2, 1, 1)
    plt.plot(shape_I_demod)
    plt.title('After Carrier removal Mix I')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(shape_Q_demod)
    plt.title('After Carrier removal Mix Q')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.grid()

    X_f3 = abs(sf.fft(shape_I_demod))
    l3 = np.size(shape_I_demod)
    fr3 = (Fs/2)*np.linspace(0, 1, int(l3/2))
    xl_m3 = (2/l3)*abs(X_f3[0:np.size(fr3)])

    plt.figure(13)
    plt.plot(fr3/1e6, 20*np.log10(xl_m3))
    plt.title('Demodulator Mixer Output')
    plt.xlabel('Frequency(MHz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()
```

### Low pass Filter

```python
# Low Pass Filter (Butterworth)

fc = 1e6

[shape_I_demod_filt, W, h] = QAM.LPF(shape_I_demod, fc, Fs)
[shape_Q_demod_filt, W, h] = QAM.LPF(shape_Q_demod, fc, Fs)

if plot == 1:
    plt.figure(14)
    plt.subplot(3, 1, 1)
    plt.plot(W, 20*np.log10(h))
    plt.title('Filter Freq. Response')
    plt.xlabel('Frequency(Hz)')
    plt.ylabel('Magnitute(dB)')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.plot(shape_I_demod_filt)
    plt.title('Filtered Signal I')
    plt.tight_layout()
    plt.ylabel('Amplitude')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.plot(shape_Q_demod_filt)
    plt.title('Filtered Signal Q')
    plt.tight_layout()
    plt.ylabel('Amplitude')
    plt.grid()
```

### Mathched Filter

```python

# Matched Filter

signal_I = QAM.MatchedFil(shape_I_demod_filt, template)
signal_Q = QAM.MatchedFil(shape_Q_demod_filt, template)



if plot == 1:
    plt.figure(15)
    plt.subplot(3, 1, 1)
    plt.plot(a,template)
    plt.title('Template for Matched filter')
    plt.tight_layout()
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.plot(signal_I)
    plt.title('Signal I')
    plt.tight_layout()
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.plot(signal_Q)
    plt.title('Signal Q')
    plt.tight_layout()
    plt.grid()
```

### Down sampling

```python

# Sampling - Gardner Algorithm

symbols_I = QAM.DownSamp(signal_I, len(data), K)
symbols_Q = QAM.DownSamp(signal_Q, len(data), K)

if plot == 1:
    plt.figure(16)
    plt.subplot(2, 1, 1)
    plt.stem(symbols_I)
    plt.title('Demodulated Simbols I')
    plt.grid()
    plt.subplot(2, 1, 2)
    plt.stem(symbols_Q)
    plt.title('Demodulated Simbols Q')
    plt.grid()
    plt.tight_layout()

    plt.figure(17)
    plt.scatter(mapI, mapQ)
    plt.title('Constelation TX')
    plt.grid()
    plt.figure(18)
    plt.scatter(symbols_I, symbols_Q)
    plt.title('Constelation RX')
    plt.grid()
    plt.tight_layout()
```

### Demodulation

```python


# Demodulator

data_demod = QAM.demodulator16QAM(symbols_I, symbols_Q, len(data))

error = 0.0

for i in range(len(data)):
    if data_demod[i] != data[i]:
        error = error + 1.0

BER = (error*100)/(len(data))

print("\n\n\n")

print("BER:", BER, "%")

print("Phase Offset Through AWGN Channel:", delta_phase/pi, "pi rad")
print("Frequency Offset AWGN Channel:", delta_freq, "Hz")

print("\n\n\n")

#print("demodulated Data",data_demod)
with open("DemodulatedBits.txt", "w", newline="") as file:
    for i in range(len(data_demod)):
        file.write(str(data_demod[i]))

#print("Return",BER/100)
```
